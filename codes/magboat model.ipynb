```python
lw = 0.33
Uh = 0.71
fp = 3.78
g = 9.81

Uc = np.sqrt(g*lw)
tc = lw/Uc
A = 0.035

Sp = 20.4e-4
c = np.sqrt(Sp)

Sta = fp*A/Uc
Stc = fp*c/Uc

print(Sta, Stc, Uc)
```

```python
Sta = 0.07
Stc = 0.1

sigma = 0.30
M = 2.5*(1 + 7/4*sigma)
Lambda = 1/8
l = 1/8202

Cd0 = 0.02

p = (Stc, Sta, M, Lambda, l, Cd0)
```

```python
def swimming(t, y, Stc, Sta, M, Lambda, l, Cd0):
    x = y[0]
    v = y[1]
    
    Ur = v - np.pi*Sta*np.sin(2*np.pi/l*Stc*t)
    
    dxdt = v
#     dvdt = 1/M*(Lambda*np.pi**2*Sta**2*(1 - 3*np.cos(4*np.pi*Stc/l*t))*Ur*np.abs(Ur) - Cd0*v**2)

    dvdt = 1/M*(Lambda*(-np.pi**2*Sta**2*np.cos(2*np.pi*Stc*t)**2 +\
                       2*np.pi**2*Sta**2*np.sin(2*np.pi*Stc*t)**2 + v*np.pi*Sta*np.sin(2*np.pi*Stc*t)) - Cd0*v**2)
       
    return [dxdt, dvdt]
    
```

```python
x0 = [0.0, 0.0]
tf = 50.0
t = np.arange(0.0, tf, 0.001)

sol = solve_ivp(swimming, (0.0, tf), x0, args=p, method='RK45', t_eval=t)

t = sol.t
xi = sol.y[0,:]
v = sol.y[1,:]
[_, a] = swimming(t, xi, Stc, Sta, M, Lambda, l, Cd0)

lDot = np.pi*Sta*np.sin(2*np.pi/l*Stc*t)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(t, xi, t, v, t, a)
# ax.set_xlim([0, 1])
plt.xlabel(r'$t$'),
# plt.ylabel(r'$\xi$')
plt.legend([r'$\xi$ (t)', r'$\dot{\xi}$ (t)', r'$\ddot{\xi}$ (t)'])

plt.show()
```
