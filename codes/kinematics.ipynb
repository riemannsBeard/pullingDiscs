```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = 1/0.3
alpha0 = np.deg2rad(30)
phi = 0
y0 = 0

N = 4

y = np.linspace(-0.2, 0.2, N)
t = np.linspace(0, N*T, 512)
tm = np.arange(0, N*T, T/8)


x = 1/2*np.cos(2*np.pi*t/T)
xm = 1/2*np.cos(2*np.pi*tm/T)
alpha = alpha0*np.cos(2*np.pi*t/T + phi)
```

```python
tm = np.tile(tm,(N,1)).T
```

```python
ym = y[np.newaxis,:] + xm[:,np.newaxis]
plt.plot(tm.T, ym.T, 'k.-')
plt.plot(t, x)
plt.show()
```
